LinearRegression with Boston dataset

```python
#load library
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np

#load linear model 
from sklearn.linear_model import LinearRegression

#load boston dataset
boston = datasets.load_boston()

#view boston dataset
print("----------------------------------------------data view--------------------------------------------------------")
print("----------------------keys---------------------")
print(boston.keys())
print("------------------data shape-------------------")
print(boston.data.shape)
print("----------------feature names------------------")
print(boston.feature_names)

print("---------------------bostonData.head() without feature--------------------------")
bostonData = pd.DataFrame(boston.data)
print('datasets.load_boston() type:',type(boston.data))
print('bostonData(pd.DataFrame) type:',type(boston.data))
print(bostonData.head())

print("---------------------bostonData.head() with feature--------------------------")
bostonData.columns = boston.feature_names
print(bostonData.head())

print("---------------------bostonData.head() with target--------------------------")
bostonData['PRICE'] = boston.target
print(bostonData.head())
print("----------------------------------------------data view--------------------------------------------------------")

print("--------------------------------------------linear training----------------------------------------------------")
#linear training
data_x = bostonData.drop('PRICE',axis = 1)   #boston.data
LinearModel = LinearRegression()             #set model
LinearModel.fit(boston.data, boston.target)  #LinearModel.fit(boston.data,boston.target)

print('LinearRegression w:',LinearModel.coef_)
print('LinearRegression b:',LinearModel.intercept_)
plt.scatter(bostonData.RM,bostonData.PRICE)
plt.show()
```

# Creating a Simulated Dataset（模拟数据集）
You need to generate a dataset of simulated dataset

when we want a dataset designed to be used with linear regression, make_regression is a good choice:

```python
#load library
from sklearn.datasets import make_regression 
import matplotlib.pyplot as plt
#generate features matrix, target vector and the true coefficients
features,target,coefficients = make_regression(n_samples=200,
                                               n_features=1,
                                               n_informative=3,
                                               noise=20,
                                               coef=True,
                                               random_state=1)
#view feature matrix and target vector
print('Feature matrix:\n',features[:3])
print('Target Vector:\n ',target[:3])
plt.scatter(features,target)
plt.plot(features,features*coefficients,color='red')
plt.xticks(())
plt.yticks(())
plt.show
```

classification

```python
#load library
%matplotlib
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
#generate features matrix, target vector and the true coefficients
features,target = make_classification(n_samples=500,
                                      n_features=2,
                                      #n_informative=0,
                                      n_redundant=0,
                                      n_classes=3,
                                      n_clusters_per_class=1,
                                      #weights=[.25,.75],
                                      #random_state=1
                                     )
#view feature matrix and target vector
print('Feature matrix:\n',features[:3])
print('Target Vector:\n ',target[:3])
plt.scatter(features[:,0],features[:,1],c=target)
plt.show()
```

clustering techniques:

```python
#load library
%matplotlib
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
#generate features matrix, target vector and the true coefficients
features,target = make_blobs(n_samples=400,
                             n_features=2,
                             centers=3,
                             shuffle=True, 
                             random_state=1)
#view feature matrix and target vector
print('Feature matrix:\n',features[:3])
print('Target Vector:\n ',target[:3])
plt.scatter(features[:,0],features[:,1],c=target)
plt.show()
```

make_gaussian_quantiles生成分组多维正态分布

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_gaussian_quantiles
features,target = make_gaussian_quantiles(n_samples=1000,
                                         n_features=2,
                                         n_classes=3,
                                         mean=[1,2],
                                         cov=2
                                        )
plt.scatter(features[:,0],features[:,1],marker='o',c=target)
```
